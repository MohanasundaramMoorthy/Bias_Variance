```python
import numpy as np
import matplotlib.pyplot as plt

# Creating our dataset
# linspace gives : Return evenly spaced numbers over a specified interval. (start,stop, number of elements)
t = np.linspace(0,1,12)
h = np.array([1.5,1.6,1.7,2.731,2.35,2.25,2.3,1.88,1.6,1.79,1.07,0.21])
plt.scatter(t,h)
```

```python
# Trying to fit the above graph using Linear Regression. i.e Fitting using a Linear equation(degree =1) to show High Bias
# This forms part of the old polynomial API.
#  numpy.ployfit(x,y)
#  xarray_like, shape (M,). x-coordinates of the M sample points (x[i], y[i]).
# yarray_like, shape (M,) or (M, K)
# y-coordinates of the sample points. Several data sets of sample points sharing the same x-coordinates can be fitted at once by passing in a 2D-array that contains one dataset per column.
p1 = np.polyfit(t,h, 1)
#Here (t,h,1) Here 1 is the polynomial
plt.scatter(t,h)
# np.polyval(x,y) x: polynomial coefficients are given in decreasing order of powers ,y : An array of numbers, c='r' colour is red
plt.plot(t,np.polyval(p1,t),c='r')
plt.xlabel('t(s)')
plt.ylabel('h(m)')
plt.title('Linear Model is underfit')
# plt.legend()


```

```python
# Fitting using a polynomial model (degree =10) to show high variance
p2 = np.polyfit(t,h,10)
plt.scatter(t,h)
plt.plot(t,np.polyval(p2,t),c='r')
plt.xlabel('t(s)')
plt.ylabel('h(m)')
plt.title('Polynomial model is overfit')
plt.legend()
```

```python
# Fitting using a Polynomial model(degree =5) to show low bias and low variance
p3 = np.polyfit(t,h,2)
plt.scatter(t,h)
plt.plot(t,np.polyval(p3,t),c='r')
plt.xlabel('t(s)')
plt.ylabel('h(m)')
plt.title('Quadratic model is Perfect with 3')

```
