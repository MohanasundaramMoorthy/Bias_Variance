```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("onlinefoods.csv")
df
```

```python
data = df
```

```python
data = data.drop(['Marital Status','Educational Qualifications','latitude','longitude','Output','Unnamed: 12'],axis =1 )
data
```

```python
# Encoder
label_encoder = preprocessing.LabelEncoder()
data['Feedback'] = label_encoder.fit_transform(data['Feedback'])
data['Feedback'].unique()
data
```

```python
label_encoder = preprocessing.LabelEncoder()
data['Gender'] = label_encoder.fit_transform(data['Gender'])
data['Gender'].unique()
data
```

```python
label_encoder = preprocessing.LabelEncoder()
data['Monthly Income'] = label_encoder.fit_transform(data['Monthly Income'])
data['Monthly Income'].unique()
data
```

```python
label_encoder = preprocessing.LabelEncoder()
data['Occupation'] = label_encoder.fit_transform(data['Occupation'])
data['Occupation'].unique()
data
```

```python
data.isna().sum()
```

```python
y = data["Feedback"]
data = data.drop('Feedback',axis=1)
data
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(data,y,test_size=0.2,random_state=0)

model= LinearRegression()
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
pred
```

```python
pred1 = model.predict(x_train)
pred1
```

```python
# finding variance using np
variance = np.var(pred)
variance
```

```python
SSE = np.mean((np.mean(pred1)-y)**2)
Bias = SSE-variance
Bias
```

```python
p1 = np.polyfit(pred1,pred,1)
plt.scatter(pred1,pred)
plt.plot(np.polyval(p1,pred1),c='r')
```
